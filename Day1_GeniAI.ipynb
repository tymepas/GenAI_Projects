```python
# It is the process of breaking text into smaller units, called tokens.
# Tokens can be words, characters, sentencces
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
```

```python
word = "NLP is fun"
word = word.lower() ## convert into lower case
tokens = word_tokenize(word) ## covert into tokenization
print(tokens)
```

```python
doc = ["I love NLP, NLP is great",
       "NLP is fun, I Code NLP in python",
       "I love python and NLP"]
print(doc)
```

```python
import pandas as pd
df = pd.DataFrame({'text': doc})
df
```

```python
df['text'] = df['text'].str.lower() #lowring the case of the text
df['tokens'] = df['text'].apply(word_tokenize) #tokenziation
df
```

```python
  ## Calculating the frequency of the text

from collections import Counter
df['freq'] = df['tokens'].apply(lambda tokens: Counter(tokens))
df
```

```python
 ## Stopwords are common words in a language that usually carry little meaning and information
 ## like -the, is, and, I, in etc
 ## removing these words helps you reduce noise in the data

 from nltk.corpus import stopwords
 nltk.download('stopwords')
```

```python
st = set(stopwords.words('english'))
st
```

```python
df['clean_tokens'] = df['tokens'].apply(lambda tokens: [word for word in tokens if word not in st])
df
```

```python
df['clean_freq'] = df['clean_tokens'].apply(lambda tokens: Counter(tokens))
df
```

```python
import string

df['Cleaned'] = df['clean_tokens'].apply(lambda tokens: [word for word in tokens if word not in string.punctuation])
df
```

```python
df = df[['text', 'clean_tokens', 'Cleaned']]
df
```

```python
df['clean_freq'] = df['Cleaned'].apply(lambda tokens: Counter(tokens))
df
```

```python
df['clean_text'] = df['Cleaned'].apply(lambda tokens: ' '.join(tokens))
df
```

```python
# Count Vectorization

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['clean_text'])
X
```

```python
mat_df = pd.DataFrame(X.toarray(), columns = vectorizer.get_feature_names_out())
mat_df
```

```python
mat_df.index = df.index
mat_df
```
