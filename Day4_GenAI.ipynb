```python
## Use Question and Answering Bot

import pandas as pd
data = {'Country': ['India', 'USA', 'China', 'Brazil', 'Russia', 'Japan'],
        'Population': [1300, 340, 1700, 20, 450, 220],
        'Revenue': [820, 670, 560, 220, 440, 230]}
df = pd.DataFrame(data)
print(df)
```

```python
table_text = df.to_markdown(index=True)
print(table_text)
```

```python
print("Ask me anything about the table----------!!! (type - 'exit') to quit")
while True:
  user_input = input("Enter your query: ")
  prompt = f"""
  You are data analyst. Use the following table to answer the user queries accurately.
  {table_text}
  Question: {user_input}
  """
  response = md.generate_content(prompt)
  print("Response from the Model: ", response.text)
  if user_input.lower() == 'exit' or user_input.lower() == 'quit':
    print("Thank you for using. GoodBye!!!")
    break
```

```python
from openai import OpenAI
client = OpenAI(api_key="#############################",
                base_url= 'https://api.deepseek.com')
```

```python
response = client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": "Hello"},
    ],
    max_tokens=100,
    temperature=0.7,
    stream=False

)

print("Assistant: ", response.choice[0].message.content)
```

```python
try:
    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {"role": "system", "content": "You are a helpful assistant"},
            {"role": "user", "content": "Hello"},
        ],
        max_tokens=100,
        temperature=0.7,
        stream=False
    )
    print("Assistant:", response.choices[0].message.content)
except Exception as e:
    print("Error:", e)
```
